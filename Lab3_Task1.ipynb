```
def line():
  print("----------------------")

def menu():
  line()
  print('Menu')
  line()
  print('1. Add Item\n2. Mark as Done\n3. View List\n4. Exit')
  line()
  choice = input('Enter your choice : ')
  return choice

data = []
choice = 'random'

while choice != '4':
  choice = menu()
  if choice == '1':
    item = input('What is to be done ? ')
    data.append(item)
    print(f'Item {item} added')
  elif choice == '2':
    item = input('What is to be marked as done ? ')
    if item in data:
      data.remove(item)
      print(f'Removed item {item}')
    else:
      print(f'Could not find item {item}')
  elif choice == '3':
    print('List of to-do items : ')
    for item in data:
      print(item)
  elif choice == '4':
    print('Goodbye!')
  else:
    print('Please enter one of 1, 2, 3 or 4')
```
